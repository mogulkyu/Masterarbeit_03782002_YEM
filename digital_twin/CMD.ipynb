| Parameter | Meaning | Why it is used in this project |
|---|---|---|
| `-b 0 -e 3600` | Defines the trip generation time window from second 0 to second 3600. This means trips are generated for a 1-hour interval. | Used to create candidate routes for the weekday morning 1-hour calibration period. |
| `-p 0.05` | Sets the generation period between trips. A trip is attempted every 0.05 seconds, which creates a very large number of candidate trips. | Used to build a rich candidate route pool so that `routeSampler.py` has many possible routes to choose from. |
| `--seed 42` | Fixes the random seed so that the random trip generation is reproducible. Running the same command again gives the same result. | Useful for calibration, because results stay consistent across repeated tests. |
| `--fringe-factor 10` | Makes fringe edges, meaning edges near the network boundary, much more likely to be chosen as trip origins or destinations. | Helps create more through-traffic entering and leaving the corridor, which is important for a corridor-based network. |
| `--min-distance 300` | Requires a minimum distance of about 300 meters between trip origin and destination. | Reduces unrealistically short local trips and increases the chance that routes pass multiple counting locations. |
| `-L` | Gives more weight to edges with more lanes when choosing trip origins and destinations. | Helps prefer major roads over small side roads, which improves the realism of the candidate route set. |
| `--speed-exponent 1` | Gives preference to faster roads when selecting trip origins and destinations. A value of 1 means a moderate weighting by speed. | Helps generate candidate routes that use the main corridor and other important roads more often. |
| `--random-routing-factor 1.5` | Adds randomness to route choice so that trips with the same origin and destination can use different plausible paths. | Increases route diversity, which is important because `routeSampler.py` can only sample from the routes that already exist in the candidate file. |
| `--validate` | Checks whether generated trips and routes are valid in the network and removes or fixes invalid ones. | Ensures that the candidate route file is usable for the next calibration step with `routeSampler.py`. |

```python
from pathlib import Path
import re

IN_FILE = Path(r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\chapter 4 - build digital twin\LSTM - detector prediction\input\landau\Candidate_Routes_Trips\candidate_raw.rou.xml")
OUT_FILE = Path(r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\chapter 4 - build digital twin\LSTM - detector prediction\input\landau\Candidate_Routes_Trips\candidate_minlen3.rou.xml")

MIN_EDGES = 3

text = IN_FILE.read_text(encoding="utf-8", errors="replace")
vehicle_blocks = re.findall(r'(<vehicle\b.*?</vehicle>)', text, flags=re.DOTALL)

kept_blocks = []
removed = []

for block in vehicle_blocks:
    m = re.search(r'<route\s+edges="([^"]+)"', block)
    if not m:
        removed.append(("NO_ROUTE", 0))
        continue

    edges = m.group(1).split()
    n_edges = len(edges)

    if n_edges >= MIN_EDGES:
        kept_blocks.append(block)
    else:
        removed.append((edges, n_edges))

with open(OUT_FILE, "w", encoding="utf-8") as f:
    f.write('<?xml version="1.0" encoding="UTF-8"?>\n\n')
    f.write('<routes xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" ')
    f.write('xsi:noNamespaceSchemaLocation="http://sumo.dlr.de/xsd/routes_file.xsd">\n')
    for block in kept_blocks:
        f.write(block + "\n")
    f.write('</routes>\n')

print("Total:", len(vehicle_blocks))
print("Kept:", len(kept_blocks))
print("Removed:", len(removed))
print("Saved to:", OUT_FILE)
```
