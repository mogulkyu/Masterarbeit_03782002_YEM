```python
from pathlib import Path
import traci
import shutil

OUT_DIR = Path(r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\queuelength")
OUT_DIR.mkdir(parents=True, exist_ok=True)

scenario = "weekend_evening"

# E2 output Netedit dosyasında şu path'e yazılıyor:
original_e2_output = Path(
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\Corridor_e2_output.xml"
)

# Simülasyondan sonra buraya senaryo adıyla kopyalayacağız:
scenario_e2_output = OUT_DIR / f"queue_length_{scenario}.xml"

traci.start([
    SUMO_GUI_BIN,
    "-n", NET_FILE,
    "-r", r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\weekend_evening_sampled_warmup15.rou.xml",
    "-a",
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\traffic_mix.add.xml,"
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\Corridor_detectors_e1.add.xml,"
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\Corridor_detectors_e2.add.xml,"
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\tls_9873_weekend_evening.xml,"
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\tls_9876_weekend_evening.xml,"
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\tls_9874_morning.xml,"
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\tls_LSA9_weekend_evening.xml,"
    r"C:\Users\mogul\OneDrive\Masaüstü\mt_emre\digital_twin\simulation\tls_9869_weekend_evening.xml",

    "--begin", "0",
    "--end", "4500",
    "--start"
])

for i in range(4500):
    traci.simulationStep()

    if i % 100 == 0:
        vehicles = traci.vehicle.getIDList()
        print(f"Step: {i}")
        print(f"Number of vehicles: {len(vehicles)}")

traci.close()

# E2 output'u senaryo klasörüne kopyala
shutil.copyfile(original_e2_output, scenario_e2_output)

print("Saved queue length output:")
print(scenario_e2_output)
```
